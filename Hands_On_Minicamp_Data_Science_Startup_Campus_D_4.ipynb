## Import the Packages

```python
# As Always Import Package!
import numpy as np # powerfull packageðŸ”¥
import pandas as pd #to do the data manipulation
import matplotlib.pyplot as plt # plot ðŸ“ˆ
from sklearn.datasets import load_diabetes # load diabetes dataset
from sklearn.model_selection import train_test_split # Model Selection
from sklearn.linear_model import LinearRegression # Regression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error # Metrics
import math #Mathematical advanced formula
```

## Regression

```python
# Load the diabetes dataset
diabetes_X, diabetes_Y=load_diabetes(return_X_y=True)
```

```python
# Print the dataset
diabetes_X
```

```python
# Use only one feature
diabetes_X=diabetes_X[:,np.newaxis,2]
diabetes_X
```

```python
# Split the data
diabetes_X_train=diabetes_X[:-20]
diabetes_X_test=diabetes_X[-20:]
```

```python
# Split the targets into training/testing sets
diabetes_Y_train=diabetes_Y[:-20]
diabetes_Y_test=diabetes_Y[-20:]
```

```python
#Create the linear regression object
model=LinearRegression()
```

```python
#Train the model
model.fit(diabetes_X_train, diabetes_Y_train)
```

```python
diabetes_Y_pred=model.predict(diabetes_X_test)
```

```python
#Model evaluation
#Mean Squared Error (MSE)
print("The MSE value is: %.2f" %mean_squared_error(diabetes_Y_test, diabetes_Y_pred))

#Root MSE (RMSE)
print("The RMSE value is: %.2f" %math.sqrt(mean_squared_error(diabetes_Y_test, diabetes_Y_pred)))
# semakin besar, semakin buruk

#Coefficient of Determination (R^2)
print("The R^2 value is: %.3f" %r2_score(diabetes_Y_test, diabetes_Y_pred)) 
# seberapa besar variabel y dipengaruhi oleh variabel x

#Mean Absolute Error (MAE)
print("The MAE value is: %.2f" %mean_absolute_error(diabetes_Y_test, diabetes_Y_pred))

#Mean Absolute Percentage Error (MAPE)
print("The MAPE value is: %.2f" %mean_absolute_percentage_error(diabetes_Y_test, diabetes_Y_pred))
```

```python
#Slope and intercept

print(model.coef_)
print(model.intercept_)
```
